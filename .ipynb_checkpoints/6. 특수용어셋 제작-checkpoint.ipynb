```python
k = db.profdata.find()
RI = []
for i in k:
    j = i['ResearchInterest'].replace("       ","").split(", ")
    for t in j:
        RI.append(t.strip())
```

```python
len(RI)
```

```python
RI[100:200]
```

```python
for i in RI:
    document = {}
    document['ResearchInterest']=i
    db.ResearchInterest.insert_one(document)
```

```python
url = "https://www.unist.ac.kr/academics/undergraduate-programs/academics-at-a-glance/"
r = requests.get(url, verify = False)
text = r.text
soup = BeautifulSoup(text,'lxml')
```

```python
soup
```

```python
soup.findAll("ul")
```

```python
for i in soup.findAll("div",{"class":"cont"}):
    print(i.find('h4').getText(),i.find("a")['href'])
```

```python
for i in soup.findAll("ul",{"class":"sub_menu s1"}):
    for j in i.findAll('li'):
        print(j.getText(),j.find("a")['href'])
```
