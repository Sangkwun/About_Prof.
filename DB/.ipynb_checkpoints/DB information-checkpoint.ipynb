### db는 mongo db형태로 제작되었다.

mongodb와 pip install 을 통하여 pymongo 를 설치한 다음에 다음과 같이 제작해보도록한다.

```python
from pymongo import MongoClient
```

```python
# mongo db클라이언트와, db를 선언한다. 본 프로젝트의 db명을 aboutdepart라고 할때
client = MongoClient()
db = client.aboutdepart
```

```python
# finaldb라는 콜렉션을 호출해 보자
print(db.finaldb.find_one().keys())
db.finaldb.find_one()
```

```python
# scholarworksfront라는 콜렉션을 호출해 보자
print(db.scholarworksfront.find_one().keys())
db.scholarworksfront.find_one()
```

```python
# excel이라는 콜렉션을 호출해 보자
print(db.excel.find_one().keys())
db.excel.find_one()
```

```python
print(db.keyword.find_one().keys())
db.keyword.find_one()
```
